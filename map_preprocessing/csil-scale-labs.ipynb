```python
df_paths = pd.DataFrame(pts['paths'])
df_labs = pd.DataFrame(labs['paths'])
```

```python
spawn_displace_x = 665
spawn_displace_y = 3170

def scaleSublistToList(list):
    newList = []
    for l in list: 
        newList.append(l[0]*7) # scale by a factor of 7
        newList.append(l[1]*7)
    
    return newList
```

```python
df_paths['points'] = df_paths['points'].apply(scaleSublistToList)
df_labs['points'] = df_labs['points'].apply(scaleSublistToList)
```
